```python
import matplotlib.pyplot as plt

def plot(data, labels, fullLabel,x_values = [25,50,75,100,125,150,175,200,225]):
    # Plotting
    plt.figure(figsize=(24, 6))  # Increased width to accommodate two plots side by side
    # Plot for means
    plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
    for idx, series in enumerate(data):
        if idx < 3:
            means = [pair[0] for pair in series]  # Extract the means from the current series
            plt.plot(x_values, means, '-o', label= labels[idx])  # Plot with line and circle markers
    plt.title('Means over Time for Multiple Series')
    plt.xlabel('Time Step')
    plt.ylabel('Mean')
    plt.legend()  # Show legend with the labels

    # Plot for variances
    plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
    for idx, series in enumerate(data):
        if idx < 3:
            variances = [pair[1] for pair in series]  # Extract the variances from the current series
            plt.plot(x_values, variances, '-o', label= labels[idx])  # Plot with line and circle markers
    plt.title('standard deviations over Time for Multiple Series')
    plt.xlabel('Time Step')
    plt.ylabel('STD')
    plt.legend()  # Show legend with the labels

    # Show plot
    plt.tight_layout()  # Adjust the layout to not overlap

    plt.suptitle(fullLabel)

    plt.show()
```

```python
def plot4d(data, labels, fullLavel, n=4,   x_values = [25,50,75,100,125,150,175,200,225]):
    plt.figure(figsize=(24, 6))  # Increased width for clarity

    # Plot for means
    plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot for means
    for idx, series in enumerate(data):
        for dimension in range(n):  # Loop through each dimension
            means = [pair[0][dimension] for pair in series]  # Extract the means for the current dimension
            print(means)
            plt.plot(x_values, means, '-o', label=f'{labels[idx]} Dim {dimension + 1}')  # Plot with line and circle markers
    plt.title('Means (4 Dimensions) over Time')
    plt.xlabel('Time Step')
    plt.ylabel('Mean')
    plt.legend()  # Show legend with the labels

    # Plot for variances
    plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot for variances
    for idx, series in enumerate(data):
        for dimension in range(n):  # Loop through each dimension
            variances = [pair[1][dimension] for pair in series]  # Extract the variances for the current dimension
            plt.plot(x_values, variances, '-o', label=f'{labels[idx]} Dim {dimension + 1}')  # Plot with line and circle markers
    plt.title('Variances (4 Dimensions) over Time')
    plt.xlabel('Time Step')
    plt.ylabel('Variance')
    plt.legend()  # Show legend with the labels

    # Add overall figure title
    plt.suptitle(fullLavel, fontsize=16)

    # Show plot
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Adjust layout to make room for the figure title
    plt.show()
print(all_rleft0)
print(all_rright1)
plot4d(all_rleft0, labels, "left action coef")
plot4d(all_rright1, labels, "right action coef")
```
