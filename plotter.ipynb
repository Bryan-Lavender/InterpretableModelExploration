```python
import numpy as np
import matplotlib.pyplot as plt
state_diffs = np.load('runs/state_diffs.np.npy')
action_diffs = np.load('runs/action_diffs.np.npy')
reward_diffs = np.load('runs/reward_diffs.np.npy')
MainPolicy_diffs = np.load('runs/MainPolicyDiff.np.npy')
```

```python
x_values = range(1000)  # This creates a range from 0 to 999
rsum_state = []
rsum_action = []
rsum_reward = []
rsum_mainPolicy = []
for i in range(0,1000):
    if i == 0:
        rsum_state.append(state_diffs[0])
        rsum_action.append(action_diffs[0])
        rsum_reward.append(reward_diffs[0])
        rsum_mainPolicy.append(MainPolicy_diffs[0])
    else:
        rsum_state.append(rsum_state[i-1] + state_diffs[i])
        rsum_action.append(rsum_action[i-1] + action_diffs[i])
        rsum_reward.append(rsum_reward[i-1] + reward_diffs[i])
        rsum_mainPolicy.append(rsum_mainPolicy[i-1] + MainPolicy_diffs[i])

# Create the plot
plt.figure(figsize=[12, 6])  # Optional: Adjusts the size of the figure
plt.plot(x_values, state_diffs, label='state')  # Plot the first line
plt.plot(x_values, action_diffs, label='action')  # Plot the second line
plt.plot(x_values, reward_diffs, label='reward')  # Plot the third line
plt.plot(x_values, MainPolicy_diffs, label='policydist')

# Adding labels and title
plt.title('Line plots of the three datasets')
plt.xlabel('Index')
plt.ylabel('Value')

# Show legend to identify each line
plt.legend()

# Show grid for better readability
plt.grid(True)

# Display the plot
plt.show()
```
