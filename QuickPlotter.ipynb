```python
import pandas as pd
import json
import matplotlib.pyplot as plt
def get_data(file_name):
    data = []

    # Open the file and read each line
    with open(file_name, 'r') as file:
        for line in file:
            # Convert the line to a dictionary
            dictionary = json.loads(line.strip())  # Strip removes extra spaces or newlines
            data.append(dictionary)

    # Now, `data` is a list of dictionaries
    return data
Domain = "_bipedal"
CartpoleBadMaxAvg = "inverting/"+Domain+"/metrics/FalseMax_avgLRP"
CartpoleBadMaxAll = "inverting/"+Domain+"/metrics/FalseMax_allLRP"
CartpoleBadVarWeight = "inverting/"+Domain+"/metrics/FalseVar_weightedLRP"
CartpoleBaseline = "inverting/"+Domain+"/metrics/FalseBaselineFD"

CartpoleGoodMaxAvg = "DTMets/"+Domain+"/Max_avg40"
CartpoleGoodMaxAll = "DTMets/"+Domain+"/Max_all40"
CartpoleGoodVarWeight = "DTMets/"+Domain+"/Var_Weighted40"

CartpoleBadMaxAvg = get_data(CartpoleBadMaxAvg)
CartpoleBadMaxAll = get_data(CartpoleBadMaxAll)
CartpoleBadVarWeight = get_data(CartpoleBadVarWeight)

CartpoleBaseline = get_data(CartpoleBaseline)

CartpoleGoodMaxAvg = get_data(CartpoleGoodMaxAvg)
CartpoleGoodMaxAll = get_data(CartpoleGoodMaxAll)
CartpoleGoodVarWeight = get_data(CartpoleGoodVarWeight)
```

```python
def getMetric(listobj, metric):
    re = []
    for i in listobj:
        re.append(i[metric])
    return re

getMetric(CartpoleBadMaxAll, "Policy_Captured")
```

```python
import numpy as np
things = [CartpoleBaseline,CartpoleBadMaxAll, CartpoleGoodMaxAll,  CartpoleBadMaxAvg, CartpoleGoodMaxAvg,CartpoleBadVarWeight, CartpoleGoodVarWeight]
data = [getMetric(thing, "Uniform_Captured") for thing in things]
for i in data:
    print(np.mean(i), np.var(i))
# Labels for the data
labels = ['Baseline', 'Eq. 5 Max all', 'Eq. 4 Max all', 'Eq. 5 Max avg', 'Eq. 4 Max avg', 'Eq. 5 Var Weighted', 'Eq. 4 Var Weighted']

# Create the violin plot
plt.violinplot(data)

# Add custom x-tick labels
plt.figure(figsize=(15, 6))

# Create the violin plot
colors = ['blue', 'red', 'green', 'red', 'green', 'red', 'green']

parts = plt.violinplot(data, showmeans=False, showmedians=True)

# Customize the colors for each violin
for i, pc in enumerate(parts['bodies']):
    pc.set_facecolor(colors[i])  # Set the face color based on your specification
    pc.set_edgecolor('black')    # Optional: Set the edge color
    pc.set_alpha(0.8)            # Set the transparency level

# Add custom x-tick labels
plt.xticks(ticks=range(1, len(data)+1), labels=labels)

# Label the axes
plt.xlabel('Equation used')
plt.ylabel('MSE')
plt.title("Uniform Capture using FD in Bipedal Walker")
# Show the plot
plt.show()
```

```python
import pandas as pd
import json
import matplotlib.pyplot as plt
def get_data(file_name):
    data = []

    # Open the file and read each line
    with open(file_name, 'r') as file:
        for line in file:
            # Convert the line to a dictionary
            dictionary = json.loads(line.strip())  # Strip removes extra spaces or newlines
            data.append(dictionary)

    # Now, `data` is a list of dictionaries
    return data
Domain = "acrobot"
dd = "Cartpole"
Baseline = "The100/"+Domain+"/metrics/TrueBaselineFD"
max_avgFD = "The100/"+Domain+"/metrics/TrueMax_avgFD"
singleFD = "The100/"+Domain+"/metrics/FalseVar_weightedFD"
Baseline_single = "The100/"+Domain+"/metrics/FalseBaselineFD"

singleFD = get_data(singleFD)
Baseline = get_data(Baseline)
max_avgFD = get_data(max_avgFD)
Baseline_single = get_data(Baseline_single)
```

```python
import numpy as np
import scipy.stats as stats
import pandas as pd
things = [Baseline_single, singleFD,Baseline, max_avgFD]
data = [getMetric(thing, "Policy_Captured") for thing in things]
for i in data:
    print(np.mean(i), np.var(i))
# Labels for the data
labels = ['Baseline Single', 'Baseline Multi', 'FD single', 'FD multi']
p_value_matrix = pd.DataFrame(np.ones((len(data), len(data))), columns=labels, index=labels)

# Perform pairwise t-tests
for i in range(len(data)):
    for j in range(i+1, len(data)):
        # Calculate the p-value
        t_stat, p_value = stats.ttest_ind(data[i], data[j], equal_var=False)  # Welch's t-test
        # Store the p-value in the matrix
        p_value_matrix.iloc[i, j] = p_value
        p_value_matrix.iloc[j, i] = p_value  # Symmetric matrix
print(p_value_matrix)
# Display the matrix with the p-values
#tools.display_dataframe_to_user(name="P-Value Matrix", dataframe=p_value_matrix)
# Create the violin plot
plt.violinplot(data)

# Add custom x-tick labels
plt.figure(figsize=(15, 6))

# Create the violin plot
colors = ['blue', 'purple', 'blue', 'green']

parts = plt.violinplot(data, showmeans=False, showmedians=True)

# Customize the colors for each violin
for i, pc in enumerate(parts['bodies']):
    pc.set_facecolor(colors[i])  # Set the face color based on your specification
    pc.set_edgecolor('black')    # Optional: Set the edge color
    pc.set_alpha(0.8)            # Set the transparency level

# Add custom x-tick labels
plt.xticks(ticks=range(1, len(data)+1), labels=labels)

# Label the axes
plt.xlabel('Method')
plt.ylabel('MSE')
plt.title(f'{dd} Uniform Capture using Multi-Trees')
# Show the plot
plt.show()
```
